```python
with open("bacillus_brevis.txt", "r") as fr:
    dna_list = fr.read()
    dna = ''.join(dna_list.split("\n"))
    print(len(dna))
genetic_code = {"UUU":"F", "UUC":"F", "UUA":"L", "UUG":"L",
    "UCU":"S", "UCC":"S", "UCA":"S", "UCG":"S",
    "UAU":"Y", "UAC":"Y", "UAA":"STOP", "UAG":"STOP",
    "UGU":"C", "UGC":"C", "UGA":"STOP", "UGG":"W",
    "CUU":"L", "CUC":"L", "CUA":"L", "CUG":"L",
    "CCU":"P", "CCC":"P", "CCA":"P", "CCG":"P",
    "CAU":"H", "CAC":"H", "CAA":"Q", "CAG":"Q",
    "CGU":"R", "CGC":"R", "CGA":"R", "CGG":"R",
    "AUU":"I", "AUC":"I", "AUA":"I", "AUG":"M",
    "ACU":"T", "ACC":"T", "ACA":"T", "ACG":"T",
    "AAU":"N", "AAC":"N", "AAA":"K", "AAG":"K",
    "AGU":"S", "AGC":"S", "AGA":"R", "AGG":"R",
    "GUU":"V", "GUC":"V", "GUA":"V", "GUG":"V",
    "GCU":"A", "GCC":"A", "GCA":"A", "GCG":"A",
    "GAU":"D", "GAC":"D", "GAA":"E", "GAG":"E",
    "GGU":"G", "GGC":"G", "GGA":"G", "GGG":"G"}

def reverse_complement(dna):
    
    rc = []
    for base in dna[::-1]:
        if base == "G":
            rc.append("C")
        if base == "C":
            rc.append("G")
        if base == "T":
            rc.append("A")
        if base == "A":
            rc.append("T")
    return ''.join(rc)

def return_proteins(rna, protein):
    rna_sequences = []
    for i in range(0, len(rna) - 3*len(protein)):
        matched = []
        rna_substring = rna[i:i+3*len(protein)]
        for j in range(0, len(rna_substring)-2, 3):
            aa = genetic_code[rna_substring[j:j+3]]
            matched.append(aa)
        matched_protein = ''.join(matched)
        # print(matched_protein)
        if matched_protein == protein:
            rna_sequences.append(rna_substring)
    return rna_sequences
```
